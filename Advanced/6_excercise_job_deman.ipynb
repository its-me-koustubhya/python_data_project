```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\singh\Documents\DATA ANALYTICS BOOTCAMP\data_jobs.csv')

# converting the job_posted_date object to datetime
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df.info()
```

*Copy()*

```python
df_India = df[df['job_country'] == 'India'].copy()
df_India.info()
```

```python
df_India['job_posted_month'] = df_India['job_posted_date'].dt.strftime('%B')
df_India
```

```python
df_India_pivot = df_India.pivot_table(index='job_posted_month', columns = 'job_title_short' ,aggfunc = 'size')
```

```python
df_India_pivot.reset_index(inplace=True)
df_India_pivot.info()
```

```python
df_India_pivot['month_no'] = pd.to_datetime(df_India_pivot['job_posted_month'], format='%B').dt.month
df_India_pivot.sort_values('month_no', inplace=True)
df_India_pivot.set_index('job_posted_month', inplace=True)
```

```python
df_India_pivot.drop(columns = 'month_no', inplace=True)
```

```python
df_India_pivot
```

```python
top_3_jobs = df_India['job_title_short'].value_counts().head(3).index
df_India_pivot[top_3_jobs].plot(kind='line')
plt.title('Monthly job posting for top data jobs in India')
plt.xlabel('Month')
plt.ylabel('No. of Job posted')
plt.legend()
plt.show()
```
