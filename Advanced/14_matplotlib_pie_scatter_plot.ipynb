**Pie charts are good for two/three variable only**

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

df = pd.read_csv(r'C:\Users\singh\Documents\DATA ANALYTICS BOOTCAMP\data_jobs.csv')

# converting the job_posted_date object to datetime
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df['job_skills'] = df['job_skills'].apply(lambda skill: ast.literal_eval(skill) if pd.notna(skill) else skill)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', ylabel='', title='Work from Home Status', startangle = 90, autopct = '%1.1f%%' )
```

```python
fig, ax = plt.subplots(1,3)

dict_col = {
  'job_work_from_home': 'Work From Home',
  'job_no_degree_mention': 'Job Degree Req.',
  'job_health_insurance': 'Health Insurance Offered'
}

for i, (col, header) in enumerate(dict_col.items()):
  ax[i].pie(df[col].value_counts(), startangle = 90, autopct = '%1.1f%%', labels =['False', 'True'])
  ax[i].set_title(header)

fig.tight_layout()
```
